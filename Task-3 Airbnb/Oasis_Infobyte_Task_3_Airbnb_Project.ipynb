#  Outlier detection in the New York City Airbnb Open Data

**Boxplots to Detect Outliers**

A quick way to visualize outliers is with boxplots:

```python
# List of numerical columns
numerical_columns = ["price", "minimum_nights", "number_of_reviews", "reviews_per_month", "availability_365"]

# Plot boxplots
plt.figure(figsize=(15, 8))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

#Outlier Detection Methods
**Here are three common statistical methods to detect outliers:**

**(a) Z Score Mean**

```python
from scipy.stats import zscore

# Compute Z-scores
z_scores = np.abs(zscore(df[numerical_columns]))

# Set a threshold (e.g., 3 standard deviations)
threshold = 3
outliers_z = (z_scores > threshold).any(axis=1)

# Filter the outliers
df_outliers_z = df[outliers_z]
print("Number of outliers detected (Z-score):", df_outliers_z.shape[0])
```

**(b) Isolation Forest (Machine Learning-Based)**
Useful for detecting multivariate outliers.

```python
from sklearn.ensemble import IsolationForest

# Fit Isolation Forest
iso_forest = IsolationForest(contamination=0.02, random_state=42)
df["outlier_iforest"] = iso_forest.fit_predict(df[numerical_columns])

# Count detected outliers
print("Number of outliers detected (Isolation Forest):", sum(df["outlier_iforest"] == -1))
```
